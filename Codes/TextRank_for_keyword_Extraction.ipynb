TextRank for keyword Extraction

```python
#importing the basic libraries
import nltk
import networkx as nx
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string
```

```python
#downloading the packages
nltk.download('punkt')
nltk.download('stopwords')

nltk.download('punkt_tab')
```

```python
def textrank_keywords(text, topn=5, window_size=4):
    word=word_tokenize(text.lower())
    stop_words=set(stopwords.words('english'))
    words=[w for w in word if not w in stop_words and w not in string.punctuation]
    graph=nx.Graph()
    for i,word in enumerate(words):
        for j in range(i+1,i+window_size):
            if j<len(words):
              graph.add_edge(word,words[i])

    scores=nx.pagerank(graph)
    ranked_words=sorted(scores.items(),key=lambda x:x[1],reverse=True)
    keywords=[word for word,score in ranked_words[:topn]]
    return keywords
```

```python
t1="""I like to play volleyball.
      I am in 6th semester.
      I lives in India."""
```

```python
keywords=textrank_keywords(t1)
print(keywords)
```
