Ngram based next word predicetion

```python
import nltk
from nltk.util import ngrams
from collections import Counter,defaultdict
import random
```

```python
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
corpus = """
I like machine learning
I like deep learning
I enjoy machine learning
I enjoy data science
Machine learning is interesting
Deep learning is powerful
"""
```

```python
tokens=nltk.word_tokenize(corpus.lower())
```

```python
n=2
bigram=list(ngrams(tokens,n))
```

```python
bigram
```

```python
model=defaultdict(Counter)
for w1,w2 in bigram:
  model[w1][w2]+=1
```

```python
model
```

```python
def predict_nextword(word):
  if word in model:
    next_words=model[word]
    prediction=max(next_words,key=next_words.get)
    return prediction
  else :
    return "No prediction available"
```

```python
input_word="learning"
print(predict_nextword(input_word))
```
