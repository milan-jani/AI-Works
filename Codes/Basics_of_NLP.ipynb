```python
t1="""His name is Ramesh.
He works in Marwadi University.
His salary is INR 50,000.
His working time starts from 9:00AM."""
```

```python
doc2=nlp(t1)
```

```python
#NER: Named entity recognition
for ent in doc2.ents:
  print(ent.text,ent.label_)
```

```python
#pos tags
for token in doc2:
  print(token.text,"->",token.pos_)
```

```python
#stemming vs lemmatization
from nltk.stem.porter import *
```

```python
p_stem=PorterStemmer()
```

```python
text="I studied the subject named AI and then meeting Mr.Rohan tomorrow in the meeting"
```

```python
for word in text.split():
  print(word,"->",p_stem.stem(word))
```

```python
#lemmatization
for token in nlp(text):
  print(token.text,"->",token.pos_,"->", token.lemma_)
```
